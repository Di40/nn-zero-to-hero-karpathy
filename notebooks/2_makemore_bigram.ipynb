# Loss function

```python
for w in words[:2]:
  chs = ['.'] + list(w) + ['.']
  for ch1, ch2 in zip(chs, chs[1:]):
    ix1 = stoi[ch1]
    ix2 = stoi[ch2]
    prob = P[ix1, ix2]
    print(f'{ch1}{ch2}: {prob:.4f}')
```

```python
1/31.0
```

We are far from uniform distribution, which is good. We have learnt something.

Let's calculate the log likelihood of the model, which is equal to the log of the product of the probabilities.

```python
logprob = 1
for w in words[:2]:
  chs = ['.'] + list(w) + ['.']
  for ch1, ch2 in zip(chs, chs[1:]):
    ix1 = stoi[ch1]
    ix2 = stoi[ch2]
    prob = P[ix1, ix2]
    logprob = torch.log(prob)
    print(f'{ch1}{ch2}: {prob:.4f} {logprob:.4f}')
```

The closer to zero it is, the higher it's probability is. Otherwise, for unlikely samples we go towards negative infinity.

Log of a product is a sum of logs.

```python
log_likelihood = 0.0
n = 0
for w in words:
  chs = ['.'] + list(w) + ['.']
  for ch1, ch2 in zip(chs, chs[1:]):
    ix1 = stoi[ch1]
    ix2 = stoi[ch2]
    prob = P[ix1, ix2]
    logprob = torch.log(prob)
    log_likelihood += logprob
    n += 1
    # print(f'{ch1}{ch2}: {prob:.4f} {logprob:.4f}')
print(f'{log_likelihood=}')
nll = - log_likelihood
print(f'{nll=}')  # We want a loss function (lower is better, not the other way around).
print(f'{nll/n=}')
```

nll/n = 2.1964 is the loss of our model. Our target value is loss = 0.

Goal: Maximize likelihood of the data w.r.t. model parameters (statistical modeling).

Equivallent to maximizing the log likelihood (because log is monotonic).

Equivallent to minimizing the negative log likelihood.

log(a\*b*c) = log(a) + log(b) + log(c)

```python
log_likelihood = 0.0
n = 0
for w in ['дејанп']:
  chs = ['.'] + list(w) + ['.']
  for ch1, ch2 in zip(chs, chs[1:]):
    ix1 = stoi[ch1]
    ix2 = stoi[ch2]
    prob = P[ix1, ix2]
    logprob = torch.log(prob)
    log_likelihood += logprob
    n += 1
    print(f'{ch1}{ch2}: {prob:.4f} {logprob:.4f}')
print(f'{log_likelihood=}')
nll = - log_likelihood
print(f'{nll=}')  # We want a loss function (lower is better, not the other way around).
print(f'{nll/n=}')
```

'нп' is 0 percent likely according to our data.

In general this means that we trust the data way too much, which in this case we shouldn't. To fix this we can use smoothing.

For example, let's use ADD 1 SMOOTHING.

```python
P = (N+1).float()
P /= P.sum(dim=1, keepdim=True)

log_likelihood = 0.0
n = 0
for w in ['дејанп']:
  chs = ['.'] + list(w) + ['.']
  for ch1, ch2 in zip(chs, chs[1:]):
    ix1 = stoi[ch1]
    ix2 = stoi[ch2]
    prob = P[ix1, ix2]
    logprob = torch.log(prob)
    log_likelihood += logprob
    n += 1
    print(f'{ch1}{ch2}: {prob:.4f} {logprob:.4f}')
print(f'{log_likelihood=}')
nll = - log_likelihood
print(f'{nll=}')  # We want a loss function (lower is better, not the other way around).
print(f'{nll/n=}')
```

We still see a very high negative value for 'нп', which makes sense, but we don't get negative infinity.

Let's put everything together (not just for the first word)

```python
# Create the dataset
xs, ys = [], []  # inputs and targets
for w in words:
  chs = ['.'] + list(w) + ['.']
  for ch1, ch2 in zip(chs, chs[1:]):
    ix1 = stoi[ch1]
    ix2 = stoi[ch2]
    xs.append(ix1)
    ys.append(ix2)

xs = torch.tensor(xs)
ys = torch.tensor(ys)
num = xs.nelement()
print(f'Number of examples: {num}')
```

```python
# Initialize the network
g = torch.Generator().manual_seed(123)
W = torch.randn((VOCAB_SIZE, VOCAB_SIZE), generator=g, requires_grad=True)

# Gradient descent
for i in range(1000):

  # Forward pass
  xenc = F.one_hot(xs, num_classes=VOCAB_SIZE).float()
  logit = xenc @ W
  counts = logit.exp()
  probs = counts / counts.sum(dim=1, keepdim=True)
  loss = -probs[torch.arange(num), ys].log().mean()
  # print(loss.item())

  # Backward pass
  W.grad = None  # set the gradient to zero
  loss.backward()

  # Update
  W.data -= 0.9 * W.grad

print(loss.item())
```

After running for a while, we expect to get the same loss as before (before adding smoothing):

nll/n = 2.1964

When W is close to zero, we obtain:

logit=0; counts=1; probs=uniform distribution

Enforcing W to be near zero is equivallent to label smoothing.

# Regularization

```python
(W**2).mean()
```

When W=0, we add zero to the loss, but if W is far from zero, squaring it gives a big amount to the loss.

```python
# Initialize the network
g = torch.Generator().manual_seed(123)
W = torch.randn((VOCAB_SIZE, VOCAB_SIZE), generator=g, requires_grad=True)

# Gradient descent
for i in range(1000):

  # Forward pass
  xenc = F.one_hot(xs, num_classes=VOCAB_SIZE).float()
  logit = xenc @ W
  counts = logit.exp()
  probs = counts / counts.sum(dim=1, keepdim=True)
  loss = -probs[torch.arange(num), ys].log().mean() + 0.01 * (W**2).mean()  # Regularization loss
  # print(loss.item())

  # Backward pass
  W.grad = None  # set the gradient to zero
  loss.backward()

  # Update
  W.data -= 0.9 * W.grad

print(loss.item())
```
